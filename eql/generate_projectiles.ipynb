```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
g = 9.8  # Acceleration due to gravity (m/s^2)
time = np.linspace(0, 1, 100)  # Time from 0 to 1 second

# Adjust velocity and angle ranges
velocities = np.linspace(1, 5, 10)  # 50 velocities from 1 to 5 m/s
angles = np.linspace(10, 44, 10)  # 20 angles from 1 to 44 degrees

projectiles = []
for v in velocities:
    for theta in angles:
        theta_rad = np.radians(theta)  # Convert angle to radians
        y = v * np.sin(theta_rad) * time - 0.5 * g * time**2  # Vertical position
        x = v * np.cos(theta_rad) * time  # Horizontal position
        projectiles.append(y)

projectiles = np.array(projectiles)
```

```python
# Plot trajectories
plt.close()

plt.figure(figsize=(10, 6))

for i in range(10):
    plt.plot(time, projectiles[-i], c='k') #label=f'v={v} m/s, ={theta}')

# Customize the plot
plt.title('Projectile Trajectories')
plt.xlabel('Time (s)')
plt.ylabel('Vertical Distance (m)')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Ground line
#plt.legend(loc='upper right', fontsize='small', ncol=2)
plt.grid()
plt.show()
```
